```python
for image in images:
    plt.show(plt.imshow(np.asarray(image)))

blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")


generated_captions = []
for image in images:
    inputs = blip_processor(images=image, return_tensors="pt")
    out = blip_model.generate(**inputs)
    caption = blip_processor.decode(out[0])
    generated_captions.append(caption)
```

```python

print("Generated Captions: ", generated_captions)

clip_model = CLIPModel.from_pretrained('openai/clip-vit-base-patch32')
clip_processor = CLIPProcessor.from_pretrained('openai/clip-vit-base-patch32')

search_term = input("Enter a term to search for an image: ")

inputs = clip_processor(text=[search_term], images=images, return_tensors='pt', padding=True)
outputs = clip_model(**inputs)

logits_per_image = outputs.logits_per_image
best_match_index = logits_per_image.argmax().item()

print("Best matching caption:", generated_captions[best_match_index])
plt.show(plt.imshow(np.asarray(images[best_match_index])))
```

### Brief Report on Image Search Implementation Using BLIP and CLIP

#### Approach

My goal was to create a system that searches for images based on user input text. The system uses two advanced models: BLIP for image captioning and CLIP for matching text to images. Here is the step-by-step approach I followed:

1. **Image Downloading**: I started by downloading images from given URLs using the requests library and loaded them using the IL library.
2. **Image Captioning with BLIP**: I used the BLIP model to generate captions for each image. This model processes each image and generates descriptive text.
3. **Text-to-Image Matching with CLIP**: After generating the captions, I used the CLIP model to match the user input text with the images. The CLIP model takes both the images and the search text as input and provides a similarity score for each image-text pair.
4. **Displaying Results**: Finally, I displayed the image that best matched the userâ€™s search text along with its generated caption.

#### Challenges Faced

1. **Model Integration**: Integrating two different models (BLIP and CLIP) was challenging. Each model has its specific input requirements and outputs.
   
2. **Logits vs. Softmax**: Initially, I used the softmax function to get probabilities from the logits output by the CLIP model.
   

#### Solutions

1. **Understanding Documentation**: I spent time reading the documentation and examples for both BLIP and CLIP models. This helped me understand how to correctly prepare inputs and interpret outputs.

2. **Using Logits Directly**: Instead of using softmax, I used the logits output from the CLIP model directly.
